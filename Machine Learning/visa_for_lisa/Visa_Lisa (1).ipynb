## Prediction Algorithm - Machine Learning

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error, mean_squared_error

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier


df['Personal Loan'] = df['Personal Loan'].astype(int)

# Drop irrelevant columns (adjust as necessary)
if 'ID' in df.columns:
    df.drop(columns=['ID'], inplace=True)


X = df.drop(columns=['Personal Loan'])  
y = df['Personal Loan']  

# Normalize numerical features 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize classifiers
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Naive Bayes": GaussianNB(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

#  Train and evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate Metrics
    accuracy = accuracy_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100  # Avoid division by zero
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    # Print Evaluation Results
    print(f"{name}:")
    print(f"  🔹 Accuracy = {accuracy:.4f}")
    print(f"  🔹 MAE      = {mae:.4f}")
    print(f"  🔹 MAPE     = {mape:.2f}%")
    print(f"  🔹 RMSE     = {rmse:.4f}")
    print(classification_report(y_test, y_pred))
    print("-" * 50)

# Visualize feature importance for tree-based models
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
feature_importance = pd.Series(rf.feature_importances_, index=df.drop(columns=['Personal Loan']).columns)
feature_importance.sort_values(ascending=False).plot(kind='bar', figsize=(10, 5), title="Feature Importance (Random Forest)")
plt.show()
```

### Saving The Models

```python
import pickle

def save_loan_model(model, filename="loan_model.pkl"):
    with open(filename, 'wb') as file:
        pickle.dump(model, file)
    print(f"✅ Loan model saved successfully as '{filename}'.")

def load_loan_model(filename="loan_model.pkl"):
    try:
        with open(filename, 'rb') as file:
            model = pickle.load(file)
        print(f"✅ Loan model loaded successfully from '{filename}'.")
        return model
    except FileNotFoundError:
        print(f"❌ Error: File '{filename}' not found. Ensure the model is saved first.")
        return None
```

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# Train and store models in a dictionary
ml = {
    "DecisionTreeClassifier": DecisionTreeClassifier().fit(X_train, y_train),
    "SVC": SVC().fit(X_train, y_train),
    "LogisticRegression": LogisticRegression().fit(X_train, y_train)
}

# Save models using the save_loan_model function
save_loan_model(ml["DecisionTreeClassifier"], "DecisionTree.pkl")
save_loan_model(ml["SVC"], "SVM.pkl")
save_loan_model(ml["LogisticRegression"], "LogisticRegression.pkl")
```
